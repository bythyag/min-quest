```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct")
messages = [
    {"role": "user", "content": "There are 10 apples on an apple tree. Three fall off. Now there are X apples. What is this an example of? Answer Choices: (a) park (b) coloring book (c) garden center (d) math problem (e) gravity?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=1000, min_p = 0.8)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
